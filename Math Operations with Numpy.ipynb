## Vectors

```python
vector = np.array([1, 2, 3, 4, 5])
print("Shape of the vector:", vector.shape)
```

## Transpose

```python
vector = np.array([[1,2,3,4,5]])
print("Row Vector:", vector, "Shape: ", vector.shape)
transpose_vector = np.transpose(vector)
print("Col Vector:", transpose_vector, "Shape: ", transpose_vector.shape)
```

```python
print("Original Matrix: \n", matrix, "Shape", matrix.shape)
print("\nTranspose Matrix: \n", np.transpose(matrix), "Shape", np.transpose(matrix).shape)
```

## Statistics

```python
marks = np.random.randint(33, 100, 20)
marks
```

```python
plt.hist(marks)
plt.show()
```

```python
#Average masks of the students
np.mean(marks)
```

```python
# Median mark of the marks
np.median(marks)
```

```python
from collections import Counter

# Mode of the marks
Counter(marks).most_common(1)
```

image.png

```python
#Standard Deviation
def std(arr):
  return np.sqrt(sum((arr - np.mean(marks))**2) / len(marks))

std(marks)
```

```python
#Percentiles
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
np.percentile(ages, 50)
```

```python
#Quantile
print(f"1st Quartile {np.quantile(ages, 0.25)}")
print(f"1st Quartile {np.quantile(ages, 0.50)}")
print(f"1st Quartile {np.quantile(ages, 0.75)}")
```

```python
plt.boxplot(ages)
plt.show()
```

```python
#Veriance
def veriance(arr):
    return np.sum((arr - np.mean(arr))**2)/(len(arr)-1)

veriance(ages)
```
