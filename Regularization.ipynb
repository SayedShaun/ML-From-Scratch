- **Ridge Regression**: Use when you have many predictors with multicollinearity and want to prevent overfitting without excluding any predictors.
  
- **Lasso Regression**: Use when feature selection is crucial, and you want a sparse model by automatically excluding less important predictors.
  
- **ElasticNet Regression**: Use when you need a balance between Ridge and Lasso, dealing with multicollinearity and performing feature selection, but with flexibility in controlling the penalties.

### Ridge Regression(L2 Regularaization)

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

x, y = load_diabetes(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

r = Ridge(alpha=0.001)
r.fit(x_train, y_train)
y_pred = r.predict(x_test)
r2_score(y_test, y_pred)
```

### Lasso Regression(L1 Regularaization)

```python
from sklearn.linear_model import Lasso

l = Lasso(alpha=0.001)
l.fit(x_train, y_train)
y_pred = l.predict(x_test)
r2_score(y_test, y_pred)
```

```python
from sklearn.datasets import make_regression

x,y = make_regression(n_samples=100, 
                      n_features=1, 
                      n_informative=1, 
                      n_targets=1,
                      noise=20,
                      random_state=13)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
import matplotlib.pyplot as plt

alphas = [0,1,5,10,30]
plt.figure(figsize=(12,6))
plt.scatter(x,y)
for i in alphas:
    L = Lasso(alpha=i)
    L.fit(x_train,y_train)
    plt.plot(x_test,L.predict(x_test),label='alpha={}'.format(i))
plt.legend()
plt.show()
```

### ElasticNet Regression

```python
from sklearn.linear_model import ElasticNet

e = ElasticNet(alpha=0.001, l1_ratio=0.005)
e.fit(x_train, y_train)
y_pred = e.predict(x_test)
r2_score(y_test, y_pred)
```
